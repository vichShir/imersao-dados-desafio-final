# Planejamento do Projeto de Drug Discovery

### Processo de Data Science
A seguir so apresentadas a definio das **etapas** do processo de Data Science empregadas neste projeto:
01. Levantamento das perguntas
    * Definio de perguntas a serem respondidas.
02. Coleta dos Dados
    * Busca das bases de dados.
03. Processamento e Organizao dos Dados
    * Padronizao, limpeza, processamento e organizao das informaes das bases de dados.
04. Anlise de Dados/Anlise Exploratria
    * Anlise dos dados das bases a fim de averiguar suas principais caractersticas.
05. Desenvolvimento de Modelos e Algoritmos
    * Definio dos modelos e algoritmos de predio de Machine Learning (ML).
06. Visualizao dos Dados
    * Visualizao dos resultados dos modelos de ML e utilizao de mtricas para sua anlise.
07. Tomada de Deciso
    * Concluses obtidas dos resultados e de todo escopo.
